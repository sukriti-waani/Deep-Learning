# <span style="color:#0B5394;"><b>11. Convolution Operation</b></span>

The core operation in CNN is:

# <span style="color:#38761D;"><b>Convolution</b></span>

It uses:

- filters
- kernels

to scan the image.

---

# <span style="color:#0B5394;"><b>12. What is a Filter / Kernel?</b></span>

A filter is a:

# <span style="color:#38761D;"><b>Small matrix used to detect features</b></span>

### Example

```text
+1   0   -1
+2   0   -2
+1   0   -1
```

This filter detects:

# <span style="color:#38761D;"><b>Vertical Edges</b></span>

---

# <span style="color:#0B5394;"><b>13. Working of Convolution</b></span>

### Step-by-step process

1. Place filter on image
2. Multiply corresponding values
3. Add all results
4. Store output
5. Move filter
6. Repeat

The produced output is called:

# <span style="color:#38761D;"><b>Feature Map</b></span>

---

# <span style="color:#0B5394;"><b>14. Formula of Output Size</b></span>

Without padding:

## <span style="color:#38761D;"><b>Output = n - f + 1</b></span>

Where:

- `n` = input size
- `f` = filter size

### Example

Input = 6 × 6  
Filter = 3 × 3

Then:

## <span style="color:#38761D;"><b>Output = 6 - 3 + 1 = 4</b></span>

So output becomes:

```text
4 × 4
```

---

# <span style="color:#0B5394;"><b>15. Stride in CNN</b></span>

Stride means:

# <span style="color:#38761D;"><b>How many steps the filter moves at one time</b></span>

### Example

Stride = 1

→ filter moves one cell at a time

Stride = 2

→ filter jumps two cells

---

# <span style="color:#0B5394;"><b>16. Effect of Stride</b></span>

### Small Stride

- larger output
- more information preserved
- more computation

### Large Stride

- smaller output
- less computation
- some information may be lost

---

# <span style="color:#0B5394;"><b>17. Output Formula with Stride</b></span>

## <span style="color:#38761D;"><b>Output = ((n - f) / s) + 1</b></span>

Where:

- `n` = input size
- `f` = filter size
- `s` = stride

### Example

Input = 7  
Filter = 3  
Stride = 2

## <span style="color:#38761D;"><b>((7 - 3)/2) + 1 = 3</b></span>

Output size = 3

---

# <span style="color:#0B5394;"><b>18. Padding in CNN</b></span>

During convolution, output size decreases.

To solve this, we use:

# <span style="color:#38761D;"><b>Padding</b></span>

Padding means:

# <span style="color:#38761D;"><b>Adding extra pixels around image borders</b></span>

---

# <span style="color:#0B5394;"><b>19. Why Padding is Needed?</b></span>

Without padding:

- image size reduces continuously
- border information gets lost

Padding helps:

1. Preserve image size
2. Preserve edge information
3. Improve feature extraction

---

# <span style="color:#0B5394;"><b>20. Types of Padding</b></span>

### 1. Zero Padding

Add zeros around image.

### 2. Same Padding

Output size remains same as input.

### 3. Valid Padding

No padding applied.

---

# <span style="color:#0B5394;"><b>21. Output Formula with Padding</b></span>

## <span style="color:#38761D;"><b>Output = ((n - f + 2p)/s) + 1</b></span>

Where:

- `n` = input size
- `f` = filter size
- `p` = padding
- `s` = stride

---

# <span style="color:#0B5394;"><b>22. Example of Padding</b></span>

Input = 6  
Filter = 3  
Stride = 1  
Desired Output = 6

Using formula:

## <span style="color:#38761D;"><b>((6 - 3 + 2p)/1) + 1 = 6</b></span>

Solving:

## <span style="color:#38761D;"><b>p = 1</b></span>

So we add 1 layer of padding.

---

# <span style="color:#0B5394;"><b>23. Activation Function in CNN</b></span>

After convolution, activation function is applied.

Most common activation:

# <span style="color:#38761D;"><b>ReLU</b></span>

### Formula

## <span style="color:#38761D;"><b>f(x) = max(0, x)</b></span>

---

# <span style="color:#0B5394;"><b>24. Why ReLU is Used?</b></span>

ReLU helps:

1. Remove negative values
2. Add non-linearity
3. Faster training
4. Reduce vanishing gradient problem

### Example

| Input | Output |
|---|---|
| -5 | 0 |
| 3 | 3 |

---

# <span style="color:#0B5394;"><b>27. Pooling Layer</b></span>

Pooling layer reduces feature map size.

This process is called:

# <span style="color:#38761D;"><b>Downsampling</b></span>

### Purpose

1. Reduce computation
2. Reduce parameters
3. Reduce overfitting
4. Keep important information

---

# <span style="color:#0B5394;"><b>28. Types of Pooling</b></span>

### 1. Max Pooling

### 2. Mean Pooling

### 3. Min Pooling

---

# <span style="color:#0B5394;"><b>29. Max Pooling</b></span>

Max Pooling selects:

# <span style="color:#38761D;"><b>Maximum value from region</b></span>

### Example

1 2  
4 3  

Output:

## <span style="color:#38761D;"><b>4</b></span>

### Why Max Pooling?

Because strongest feature is preserved.

---

# <span style="color:#0B5394;"><b>30. Mean / Average Pooling</b></span>

Average Pooling calculates:

# <span style="color:#38761D;"><b>Average of region values</b></span>

### Example

```text
1 2
3 4
```

Output:

## <span style="color:#38761D;"><b>(1+2+3+4)/4 = 2.5</b></span>

---

# <span style="color:#0B5394;"><b>31. Min Pooling</b></span>

Min Pooling selects:

# <span style="color:#38761D;"><b>Minimum value</b></span>

from pooling region.

It is less commonly used.

---

# <span style="color:#0B5394;"><b>32. Pooling Parameters</b></span>

- filter size
- stride

### Common pooling

## <span style="color:#38761D;"><b>2 × 2 filter with stride 2</b></span>

This reduces image size significantly.

---

# <span style="color:#0B5394;"><b>59. Simple CNN Model</b></span>

```python
model = Sequential()

# First Convolution Layer
model.add(Conv2D(
    filters=32,
    kernel_size=(3,3),
    activation='relu',
    input_shape=(128,128,3)
))

# Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2)))

# Second Convolution Layer
model.add(Conv2D(
    filters=64,
    kernel_size=(3,3),
    activation='relu'
))

# Second Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2)))

# Flatten Layer
model.add(Flatten())

# Fully Connected Layer
model.add(Dense(units=128, activation='relu'))

# Output Layer
model.add(Dense(units=1, activation='sigmoid'))
```

# <span style="color:#0B5394;"><b>58. TensorFlow / Keras Implementation of CNN</b></span>

````python
### Import Libraries


import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
````

# <span style="color:#0B5394;"><b>59. Simple CNN Model</b></span>

````python
model = Sequential()

# First Convolution Layer
model.add(Conv2D(
    filters=32,
    kernel_size=(3,3),
    activation='relu',
    input_shape=(128,128,3)
))

# Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2)))

# Second Convolution Layer
model.add(Conv2D(
    filters=64,
    kernel_size=(3,3),
    activation='relu'
))

# Second Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2)))

# Flatten Layer
model.add(Flatten())

# Fully Connected Layer
model.add(Dense(units=128, activation='relu'))

# Output Layer
model.add(Dense(units=1, activation='sigmoid'))
````
