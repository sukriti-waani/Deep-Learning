# <span style="color:#0B5394;"><b>1. Introduction</b></span>

In machine learning and deep learning, a model learns by adjusting its **weights** and **biases**.

At the beginning, these values are random, so predictions are wrong.

We use a **Loss Function** to measure how wrong the prediction is.

But after finding the error, an important question comes:

**How do we reduce this error?**

The answer is:

# <span style="color:#38761D;"><b>Optimizer</b></span>

An optimizer helps the model update weights and biases in such a way that the loss becomes smaller and the model becomes better.

Its main goal is:

# <span style="color:#38761D;"><b>Minimize Loss Function / Cost Function</b></span>

---

# <span style="color:#0B5394;"><b>2. What is an Optimizer?</b></span>

An **Optimizer** is an algorithm used to improve the model by updating the model parameters such as:

- Weights
- Biases

It helps the model move toward the **minimum loss point**.

### Simple meaning:

Optimizer = weight updating method

### Example:

If the model predicts:

Actual = 10  
Predicted = 6

There is error.

The optimizer decides:

- how much weight should change
- in which direction it should change

so that next prediction becomes better.

---

# <span style="color:#0B5394;"><b>3. Why Do We Need Optimizers?</b></span>

Without an optimizer:

- the model cannot learn
- weights will remain the same
- prediction error will not reduce

### Role of optimizer:

1. Reduce loss function
2. Improve prediction accuracy
3. Update weights automatically
4. Help gradient descent work efficiently
5. Speed up convergence

### Simple idea:

Loss tells the problem  
Optimizer gives the solution

---

# <span style="color:#0B5394;"><b>4. General Weight Update Rule</b></span>

The basic formula used by optimizers is:

## <span style="color:#38761D;"><b>W(new) = W(old) - η × (∂L / ∂W)</b></span>

Where:

- `W` = weight
- `η` = learning rate
- `L` = loss function
- `∂L / ∂W` = gradient (slope)

### Meaning:

- Gradient tells how much error changes
- Learning rate tells how big the step should be

The optimizer uses both to update weights.

---

# <span style="color:#0B5394;"><b>5. Learning Rate</b></span>

Learning rate is one of the most important parts of optimization.

It controls:

# <span style="color:#38761D;"><b>How much the weights should change</b></span>

### If learning rate is too small:

- training becomes very slow
- model takes too much time

### If learning rate is too large:

- model may skip the minimum point
- training becomes unstable

### Ideal case:

Learning rate should be balanced.

### Example:

Like walking down stairs:

- too small step → very slow
- too large step → may fall
- balanced step → reach safely

---

# <span style="color:#0B5394;"><b>6. Gradient Descent Optimizer</b></span>

Gradient Descent is the most basic optimizer.

It updates weights using the **entire dataset** before one update.

### Process:

1. Take all training data
2. Calculate total cost
3. Find gradient
4. Update weights once

### Formula:

## <span style="color:#38761D;"><b>W(new) = W(old) - η × (∂J / ∂W)</b></span>

Where:

- `J` = cost function using full dataset

### Example:

Dataset = 1000 data points

Gradient Descent uses all 1000 points first, then updates weights.

---

# <span style="color:#0B5394;"><b>7. Epoch and Iteration</b></span>

These two words are very important.

### Epoch

One complete pass through the full dataset.

### Iteration

One weight update step.

### Example:

Dataset = 1000 samples

In Gradient Descent:

- 1 Epoch = 1 Iteration

because full data is used once before updating weights.

---

# <span style="color:#0B5394;"><b>8. Graph of Gradient Descent</b></span>

The cost function graph is usually bowl-shaped.

### Graph meaning:

- top = high loss
- bottom = minimum loss

Gradient Descent moves step by step toward the bottom.

### Interpretation:

The optimizer tries to reach:

# <span style="color:#38761D;"><b>Global Minimum</b></span>

where error becomes minimum.

---

# <span style="color:#0B5394;"><b>9. Advantages of Gradient Descent</b></span>

1. Convergence is stable
2. Smooth movement toward minimum
3. Works well for small datasets
4. Easy to understand
5. Easy to implement

---

# <span style="color:#0B5394;"><b>10. Disadvantages of Gradient Descent</b></span>

1. Very slow for large datasets
2. Needs high RAM
3. Resource intensive (CPU/GPU)
4. One update takes long time

This is why it becomes difficult for very large datasets.

---

# <span style="color:#0B5394;"><b>11. Stochastic Gradient Descent (SGD)</b></span>

SGD means:

# <span style="color:#38761D;"><b>Stochastic Gradient Descent</b></span>

Instead of using the full dataset, SGD updates weights using:

# <span style="color:#38761D;"><b>Only one data point at a time</b></span>

### Process:

1. Take one sample
2. Calculate loss
3. Update weights immediately
4. Move to next sample

### Example:

Dataset = 1000 samples

SGD performs:

1000 updates in 1 epoch

---

# <span style="color:#0B5394;"><b>12. Epoch and Iteration in SGD</b></span>

Dataset = 1000 samples

### In SGD:

- 1 Epoch = full dataset once
- 1000 Iterations = because update happens for every sample

### Formula idea:

More updates → faster movement

but also more noise.

---

# <span style="color:#0B5394;"><b>13. Graph of SGD</b></span>

The path of SGD looks:

# <span style="color:#38761D;"><b>Zig-Zag</b></span>

because every sample gives a slightly different gradient.

### Interpretation:

- faster updates
- noisy path
- may jump around minimum point

It does not move smoothly like Gradient Descent.

---

# <span style="color:#0B5394;"><b>14. Advantages of SGD</b></span>

1. Solves memory problem
2. Faster updates
3. Good for very large datasets
4. Less RAM required
5. Can escape local minima better

---

# <span style="color:#0B5394;"><b>15. Disadvantages of SGD</b></span>

1. Very noisy updates
2. Convergence takes more time
3. Path is unstable
4. May not reach exact minimum
5. Oscillation happens near minimum

---

# <span style="color:#0B5394;"><b>16. Quick Difference: GD vs SGD</b></span>

| Parameter | Gradient Descent | SGD |
|---|---|---|
| Data Used | Full dataset | One sample |
| Speed | Slow | Faster |
| Memory | High | Low |
| Update Frequency | Once per epoch | Every sample |
| Path | Smooth | Noisy |
| Convergence | Stable | Fast but unstable |

---

# <span style="color:#0B5394;"><b>17. Real-Life Analogy</b></span>

Imagine checking exam copies.

### Gradient Descent:

Teacher checks all 1000 answer sheets first, then gives final feedback.

### SGD:

Teacher checks one answer sheet and gives feedback immediately.

### Result:

- GD = slow but stable
- SGD = fast but noisy


---

# <span style="color:#0B5394;"><b>18. Mini Batch Gradient Descent</b></span>

Mini Batch Gradient Descent is the combination of:

- Gradient Descent
- Stochastic Gradient Descent (SGD)

It does not use:

- full dataset like GD
- one sample like SGD

Instead, it uses:

# <span style="color:#38761D;"><b>A small group of samples called Batch</b></span>

### Example:

Dataset = 100000 samples  
Batch Size = 1000

Then:

Each update uses 1000 samples only

---

# <span style="color:#0B5394;"><b>19. Batch Size</b></span>

Batch size means:

# <span style="color:#38761D;"><b>Number of training samples used in one iteration</b></span>

### Example:

Dataset = 10000 samples  
Batch size = 100

Then:

## <span style="color:#38761D;"><b>Total Iterations = 10000 / 100 = 100</b></span>

### Common batch sizes:

- 32
- 64
- 128
- 256
- 512
- 1024

The choice depends on:

- RAM
- GPU memory
- dataset size

---

# <span style="color:#0B5394;"><b>20. Working of Mini Batch SGD</b></span>

### Process:

1. Divide dataset into small batches
2. Take one batch
3. Calculate cost function
4. Find gradient
5. Update weights
6. Move to next batch

### Example:

10000 samples → divided into 100 batches

Each batch updates weights separately.

---

# <span style="color:#0B5394;"><b>21. Formula of Mini Batch SGD</b></span>

The formula is same as GD:

## <span style="color:#38761D;"><b>W(new) = W(old) - η × (∂J / ∂W)</b></span>

But here:

- `J` is calculated using only one mini batch

not the full dataset.

---

# <span style="color:#0B5394;"><b>22. Graph of Mini Batch SGD</b></span>

The path looks:

# <span style="color:#38761D;"><b>Less noisy than SGD and faster than GD</b></span>

### Interpretation:

- smoother than SGD
- faster than GD
- better convergence

It is the most commonly used basic optimizer.

---

# <span style="color:#0B5394;"><b>23. Advantages of Mini Batch SGD</b></span>

1. Faster convergence
2. Less noise than SGD
3. Efficient memory usage
4. Works well with GPUs
5. Best balance between speed and stability

---

# <span style="color:#0B5394;"><b>24. Disadvantages of Mini Batch SGD</b></span>

1. Still has some noise
2. Batch size selection is important
3. Wrong batch size may reduce performance

---

# <span style="color:#0B5394;"><b>25. SGD with Momentum</b></span>

Normal SGD moves in a zig-zag path.

This causes:

- slow convergence
- oscillation near minimum

To solve this problem, we use:

# <span style="color:#38761D;"><b>Momentum</b></span>

Momentum helps the optimizer remember the previous direction and continue moving smoothly.

### Simple meaning:

Momentum = speed + direction memory

---

# <span style="color:#0B5394;"><b>26. Real-Life Example of Momentum</b></span>

Imagine pushing a heavy ball down a hill.

At first:

- movement is slow

After some time:

- ball gains speed
- movement becomes smoother

This is called momentum.

Same idea is used in optimization.

---

# <span style="color:#0B5394;"><b>27. Formula of Momentum</b></span>

First we calculate velocity:

## <span style="color:#38761D;"><b>Vt = β × V(t-1) + (1 - β) × Gradient</b></span>

Then weight update:

## <span style="color:#38761D;"><b>W(new) = W(old) - η × Vt</b></span>

Where:

- `Vt` = velocity
- `β` = momentum term
- `η` = learning rate

Usually:

## <span style="color:#38761D;"><b>β = 0.9 or 0.95</b></span>

---

# <span style="color:#0B5394;"><b>28. Exponential Weighted Average (EWA)</b></span>

Momentum uses:

# <span style="color:#38761D;"><b>Exponential Weighted Average</b></span>

This means:

Recent gradients get more importance  
Old gradients get less importance

### Example:

## <span style="color:#38761D;"><b>Vt = βV(t-1) + (1-β)at</b></span>

This creates smoothing.

That is why movement becomes less noisy.

---

# <span style="color:#0B5394;"><b>29. Graph of Momentum</b></span>

Instead of strong zig-zag movement:

Momentum gives:

# <span style="color:#38761D;"><b>Smoother and Faster Path</b></span>

### Interpretation:

- less oscillation
- faster convergence
- reduced noise

The optimizer reaches minimum faster.

---

# <span style="color:#0B5394;"><b>30. Advantages of Momentum</b></span>

1. Reduces zig-zag movement
2. Faster convergence
3. Reduces noise
4. Better performance than normal SGD
5. Helps escape shallow local minima

---

# <span style="color:#0B5394;"><b>31. Disadvantages of Momentum</b></span>

1. Need proper beta value
2. Wrong momentum can overshoot minimum
3. Slightly more complex than SGD

---

# <span style="color:#0B5394;"><b>32. Quick Difference: GD vs SGD vs Mini Batch vs Momentum</b></span>

| Parameter | GD | SGD | Mini Batch | Momentum |
|---|---|---|---|---|
| Data Used | Full dataset | One sample | Small batch | Usually mini batch |
| Speed | Slow | Fast | Balanced | Faster |
| Noise | Very low | Very high | Medium | Low |
| Memory | High | Low | Medium | Medium |
| Convergence | Stable | Unstable | Better | Faster and smoother |

---

# <span style="color:#0B5394;"><b>33. Simple Exam Line</b></span>

Mini Batch Gradient Descent uses a small group of samples for one update and gives better balance between speed and stability.

Momentum improves SGD by reducing oscillation and making convergence faster using previous gradient direction.

---

# <span style="color:#0B5394;"><b>34. AdaGrad Optimizer</b></span>

AdaGrad means:

# <span style="color:#38761D;"><b>Adaptive Gradient Descent</b></span>

In normal Gradient Descent, the learning rate remains fixed.

But in real problems:

- some parameters need large updates
- some parameters need small updates

AdaGrad solves this by using:

# <span style="color:#38761D;"><b>Different learning rates for different parameters</b></span>

### Simple meaning:

Learning rate changes automatically

This is called:

# <span style="color:#38761D;"><b>Dynamic Learning Rate</b></span>

---

# <span style="color:#0B5394;"><b>35. Why AdaGrad is Needed</b></span>

Fixed learning rate causes problems:

### If learning rate is large:

- model may overshoot minimum

### If learning rate is small:

- training becomes very slow

AdaGrad adjusts learning rate automatically during training.

As convergence happens:

# <span style="color:#38761D;"><b>Learning rate should change</b></span>

This improves training.

---

# <span style="color:#0B5394;"><b>36. Formula of AdaGrad</b></span>

First:

## <span style="color:#38761D;"><b>Gt = Σ (∂L/∂W)²</b></span>

This stores the sum of squared gradients.

Then new learning rate:

## <span style="color:#38761D;"><b>η' = η / √(Gt + ε)</b></span>

Finally:

## <span style="color:#38761D;"><b>Wt = W(t-1) - η' × (∂L/∂W)</b></span>

Where:

- `η` = original learning rate
- `η'` = updated learning rate
- `ε` = very small value to avoid division by zero

---

# <span style="color:#0B5394;"><b>37. Working of AdaGrad</b></span>

### At beginning:

learning rate is large

### After many updates:

Gt increases

so:

## <span style="color:#38761D;"><b>η' becomes smaller</b></span>

This means:

- big steps at start
- small steps near minimum

This helps in better convergence.

---

# <span style="color:#0B5394;"><b>38. Graph of AdaGrad</b></span>

The optimizer starts fast and becomes slower near the minimum.

### Interpretation:

- strong movement at beginning
- slow careful movement near minimum

This is useful for sparse data.

---

# <span style="color:#0B5394;"><b>39. Advantages of AdaGrad</b></span>

1. Automatic learning rate adjustment
2. Good for sparse data
3. Less manual tuning needed
4. Useful in NLP and text problems

---

# <span style="color:#0B5394;"><b>40. Disadvantages of AdaGrad</b></span>

1. Learning rate becomes too small
2. Training may stop early
3. Sometimes convergence becomes too slow

This is the biggest problem of AdaGrad.

---

# <span style="color:#0B5394;"><b>41. RMSProp Optimizer</b></span>

RMSProp means:

# <span style="color:#38761D;"><b>Root Mean Square Propagation</b></span>

RMSProp improves AdaGrad.

It solves the problem:

# <span style="color:#38761D;"><b>Learning rate becoming too small</b></span>

Instead of storing all past squared gradients forever,
RMSProp stores only:

# <span style="color:#38761D;"><b>Recent gradients using moving average</b></span>

This gives better control.

---

# <span style="color:#0B5394;"><b>42. Formula of RMSProp</b></span>

First:

## <span style="color:#38761D;"><b>Sdw = β × Sdw + (1 - β)(∂L/∂W)²</b></span>

Then:

## <span style="color:#38761D;"><b>η' = η / √(Sdw + ε)</b></span>

Finally:

## <span style="color:#38761D;"><b>Wt = W(t-1) - η' × (∂L/∂W)</b></span>

Usually:

## <span style="color:#38761D;"><b>β = 0.9</b></span>

Where:

- `Sdw` = moving average of squared gradients
- `β` = smoothing factor

---

# <span style="color:#0B5394;"><b>43. Why RMSProp is Better than AdaGrad</b></span>

### AdaGrad:

stores all old gradients forever

→ learning rate becomes too small

### RMSProp:

stores only recent gradients

→ learning rate remains useful

That is why RMSProp performs better in deep learning.

---

# <span style="color:#0B5394;"><b>44. Graph of RMSProp</b></span>

The path becomes:

# <span style="color:#38761D;"><b>Smoother + Faster + Stable</b></span>

### Interpretation:

- avoids too small learning rate
- faster than AdaGrad
- better practical performance

---

# <span style="color:#0B5394;"><b>45. Advantages of RMSProp</b></span>

1. Solves AdaGrad problem
2. Better dynamic learning rate
3. Faster convergence
4. Very common in deep learning
5. Stable training

---

# <span style="color:#0B5394;"><b>46. Disadvantages of RMSProp</b></span>

1. Need beta tuning
2. Slightly complex
3. Sometimes still needs manual tuning

---

# <span style="color:#0B5394;"><b>47. Quick Difference: AdaGrad vs RMSProp</b></span>

| Parameter | AdaGrad | RMSProp |
|---|---|---|
| Learning Rate | Keeps decreasing | Controlled decrease |
| Gradient Memory | All past gradients | Recent gradients only |
| Speed | May become slow | Faster |
| Stability | Lower | Higher |
| Performance | Good for sparse data | Better for deep learning |

---

# <span style="color:#0B5394;"><b>48. Simple Exam Line</b></span>

AdaGrad uses adaptive learning rate for each parameter, but its learning rate becomes too small over time.

RMSProp improves AdaGrad by using moving average of recent gradients and gives better stable convergence.


---

# <span style="color:#0B5394;"><b>49. Adam Optimizer</b></span>

Adam means:

# <span style="color:#38761D;"><b>Adaptive Moment Estimation</b></span>

Adam is one of the most powerful and most commonly used optimizers in deep learning.

It combines the ideas of:

- Momentum
- RMSProp

That means Adam gives:

# <span style="color:#38761D;"><b>Fast convergence + Smooth movement + Dynamic learning rate</b></span>

This is why Adam is often the default optimizer.

---

# <span style="color:#0B5394;"><b>50. Why Adam is Needed</b></span>

Problems in previous optimizers:

### SGD

- noisy updates
- slow convergence

### Momentum

- improves direction but still needs tuning

### AdaGrad

- learning rate becomes too small

### RMSProp

- better learning rate but no momentum effect

Adam solves all these together.

That is why it is called:

# <span style="color:#38761D;"><b>Best practical optimizer for many deep learning problems</b></span>

---

# <span style="color:#0B5394;"><b>51. Formula of Adam Optimizer</b></span>

Adam uses two moving averages:

### First moment (Momentum)

## <span style="color:#38761D;"><b>Vdw = β₁ × Vdw + (1 - β₁)(∂L/∂W)</b></span>

### Second moment (RMSProp)

## <span style="color:#38761D;"><b>Sdw = β₂ × Sdw + (1 - β₂)(∂L/∂W)²</b></span>

### Updated learning rate:

## <span style="color:#38761D;"><b>η' = η / √(Sdw + ε)</b></span>

### Final weight update:

## <span style="color:#38761D;"><b>Wt = W(t-1) - η' × Vdw</b></span>

Where:

- `Vdw` = momentum term
- `Sdw` = squared gradient moving average
- `β₁` = momentum factor
- `β₂` = RMSProp factor

Usually:

## <span style="color:#38761D;"><b>β₁ = 0.9</b></span>

## <span style="color:#38761D;"><b>β₂ = 0.999</b></span>

---

# <span style="color:#0B5394;"><b>52. Working of Adam</b></span>

### Step 1

Use momentum to smooth gradient direction

### Step 2

Use RMSProp to control learning rate

### Step 3

Update weights using both together

### Result

# <span style="color:#38761D;"><b>Faster + Stable + Smart Optimization</b></span>

This makes training much better.

---

# <span style="color:#0B5394;"><b>53. Graph of Adam Optimizer</b></span>

Adam reaches the minimum:

# <span style="color:#38761D;"><b>Fastest and Smoothest</b></span>

### Interpretation:

- very little oscillation
- fast convergence
- stable updates
- strong practical performance

Among all optimizers, Adam is usually preferred.

---

# <span style="color:#0B5394;"><b>54. Advantages of Adam</b></span>

1. Fast convergence
2. Dynamic learning rate
3. Uses momentum smoothing
4. Very stable
5. Works well for most deep learning tasks
6. Less manual tuning needed
7. Excellent default optimizer

---

# <span style="color:#0B5394;"><b>55. Disadvantages of Adam</b></span>

1. More memory usage
2. Slightly more complex
3. Sometimes may not generalize as well as SGD in some tasks
4. Hyperparameter tuning may still be needed

---

# <span style="color:#0B5394;"><b>56. Complete Comparison Table of Optimizers</b></span>

| Optimizer | Data Used | Learning Rate | Noise | Speed | Best For |
|---|---|---|---|---|---|
| Gradient Descent | Full dataset | Fixed | Low | Slow | Small datasets |
| SGD | One sample | Fixed | High | Fast | Large datasets |
| Mini Batch SGD | Small batch | Fixed | Medium | Balanced | Most practical training |
| Momentum | Mini batch + previous direction | Fixed | Low | Faster | Faster convergence |
| AdaGrad | Adaptive | Dynamic | Medium | Slower later | Sparse data |
| RMSProp | Adaptive recent gradients | Dynamic | Low | Fast | Deep learning |
| Adam | Momentum + RMSProp | Dynamic | Very Low | Very Fast | Most deep learning tasks |

---

# <span style="color:#0B5394;"><b>57. When to Use Which Optimizer</b></span>

### Use Gradient Descent when:

- dataset is small
- stability is important

### Use SGD when:

- dataset is very large
- fast updates are needed

### Use Mini Batch SGD when:

- balanced performance is needed

### Use Momentum when:

- SGD is too noisy

### Use AdaGrad when:

- sparse data exists
- NLP problems are present

### Use RMSProp when:

- adaptive learning is needed

### Use Adam when:

- general deep learning tasks
- CNN
- RNN
- ANN
- most real-world projects

Adam is usually the first choice.

```python
# =========================================
# Graphs of Optimizers in Deep Learning
# =========================================

import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Common Cost Function
# -----------------------------
# Simple quadratic cost function:
# J(w) = w^2

w = np.linspace(-6, 6, 500)
J = w**2


# =========================================
# 1. Gradient Descent Graph (Smooth Path)
# =========================================

plt.figure(figsize=(8,5))
plt.plot(w, J, label="Cost Function J(w) = w²")

gd_points = np.array([5, 4, 3, 2, 1, 0.5, 0.2, 0])
plt.plot(gd_points, gd_points**2, marker='o', linestyle='--', label="Gradient Descent Path")

plt.title("Gradient Descent")
plt.xlabel("Weight (w)")
plt.ylabel("Cost J(w)")
plt.legend()
plt.grid(True)
plt.show()


# =========================================
# 2. SGD Graph (Zig-Zag Path)
# =========================================

plt.figure(figsize=(8,5))
plt.plot(w, J, label="Cost Function J(w) = w²")

sgd_points = np.array([5, 3.8, 4.2, 2.5, 1.8, 2.2, 0.8, 0.3, 0])
plt.plot(sgd_points, sgd_points**2, marker='o', linestyle='--', label="SGD Path")

plt.title("Stochastic Gradient Descent (SGD)")
plt.xlabel("Weight (w)")
plt.ylabel("Cost J(w)")
plt.legend()
plt.grid(True)
plt.show()


# =========================================
# 3. Mini Batch SGD Graph
# =========================================

plt.figure(figsize=(8,5))
plt.plot(w, J, label="Cost Function J(w) = w²")

mini_batch_points = np.array([5, 4.2, 3.1, 2.0, 1.2, 0.6, 0.2, 0])
plt.plot(mini_batch_points, mini_batch_points**2, marker='o', linestyle='--', label="Mini Batch Path")

plt.title("Mini Batch Gradient Descent")
plt.xlabel("Weight (w)")
plt.ylabel("Cost J(w)")
plt.legend()
plt.grid(True)
plt.show()


# =========================================
# 4. Momentum Graph (Smoother Faster Path)
# =========================================

plt.figure(figsize=(8,5))
plt.plot(w, J, label="Cost Function J(w) = w²")

momentum_points = np.array([5, 3.5, 2.2, 1.1, 0.4, 0.1, 0])
plt.plot(momentum_points, momentum_points**2, marker='o', linestyle='--', label="Momentum Path")

plt.title("SGD with Momentum")
plt.xlabel("Weight (w)")
plt.ylabel("Cost J(w)")
plt.legend()
plt.grid(True)
plt.show()


# =========================================
# 5. AdaGrad Learning Rate Graph
# =========================================

iterations = np.arange(1, 51)
learning_rate_adagrad = 1 / np.sqrt(iterations)

plt.figure(figsize=(8,5))
plt.plot(iterations, learning_rate_adagrad)

plt.title("AdaGrad Learning Rate Decrease")
plt.xlabel("Iterations")
plt.ylabel("Learning Rate")
plt.grid(True)
plt.show()


# =========================================
# 6. RMSProp Learning Rate Graph
# =========================================

learning_rate_rmsprop = 1 / np.sqrt(np.log(iterations + 1) + 1)

plt.figure(figsize=(8,5))
plt.plot(iterations, learning_rate_rmsprop)

plt.title("RMSProp Controlled Learning Rate")
plt.xlabel("Iterations")
plt.ylabel("Learning Rate")
plt.grid(True)
plt.show()


# =========================================
# 7. Adam Optimizer Graph (Fastest Path)
# =========================================

plt.figure(figsize=(8,5))
plt.plot(w, J, label="Cost Function J(w) = w²")

adam_points = np.array([5, 2.8, 1.4, 0.6, 0.2, 0.05, 0])
plt.plot(adam_points, adam_points**2, marker='o', linestyle='--', label="Adam Path")

plt.title("Adam Optimizer")
plt.xlabel("Weight (w)")
plt.ylabel("Cost J(w)")
plt.legend()
plt.grid(True)
plt.show()
```

# <span style="color:#0B5394;"><b>61. Graph Explanation for Optimizers</b></span>

The graphs help us understand how different optimizers move toward the minimum loss point.

The common cost function used is:

## <span style="color:#38761D;"><b>J(w) = w²</b></span>

This creates a bowl-shaped curve.

### Meaning of graph:

- X-axis = Weight (w)
- Y-axis = Cost Function J(w)

The lowest point:

## <span style="color:#38761D;"><b>w = 0</b></span>

is the minimum loss point.

All optimizers try to reach this minimum.

---

# <span style="color:#0B5394;"><b>62. Gradient Descent Graph Explanation</b></span>

Gradient Descent moves:

# <span style="color:#38761D;"><b>Smoothly and Slowly</b></span>

toward the minimum point.

### Graph behavior:

- uses full dataset
- stable movement
- very little noise
- takes more time

### Interpretation:

Best for small datasets where stability is more important than speed.

---

# <span style="color:#0B5394;"><b>63. SGD Graph Explanation</b></span>

SGD moves in a:

# <span style="color:#38761D;"><b>Zig-Zag Path</b></span>

because it updates using one sample at a time.

### Graph behavior:

- very noisy updates
- fast movement
- unstable near minimum
- may jump around minimum point

### Interpretation:

Good for very large datasets but less stable.

---

# <span style="color:#0B5394;"><b>64. Mini Batch Graph Explanation</b></span>

Mini Batch SGD gives:

# <span style="color:#38761D;"><b>Balanced Movement</b></span>

between GD and SGD.

### Graph behavior:

- smoother than SGD
- faster than GD
- moderate noise
- practical for deep learning

### Interpretation:

Most commonly used basic optimizer.

---

# <span style="color:#0B5394;"><b>65. Momentum Graph Explanation</b></span>

Momentum gives:

# <span style="color:#38761D;"><b>Faster and Smoother Path</b></span>

because it remembers previous direction.

### Graph behavior:

- reduced zig-zag movement
- less oscillation
- reaches minimum faster

### Interpretation:

Better than normal SGD.

---

# <span style="color:#0B5394;"><b>66. AdaGrad Graph Explanation</b></span>

AdaGrad starts:

# <span style="color:#38761D;"><b>Fast and then Slows Down</b></span>

because learning rate decreases automatically.

### Graph behavior:

- large steps at beginning
- small steps near minimum
- useful for sparse data

### Problem:

Sometimes learning rate becomes too small.

---

# <span style="color:#0B5394;"><b>67. RMSProp Graph Explanation</b></span>

RMSProp gives:

# <span style="color:#38761D;"><b>Stable Learning Rate Control</b></span>

by using only recent gradients.

### Graph behavior:

- avoids too small learning rate
- smoother than AdaGrad
- faster convergence

### Interpretation:

Very strong practical optimizer.

---

# <span style="color:#0B5394;"><b>68. Adam Graph Explanation</b></span>

Adam gives:

# <span style="color:#38761D;"><b>Fastest and Smartest Movement</b></span>

because it combines:

- Momentum
- RMSProp

### Graph behavior:

- very smooth path
- very fast convergence
- low oscillation
- strong stability

### Interpretation:

Most preferred optimizer in deep learning.
