# Multilayer Perceptron (Artificial Neural Network)

A Multilayer Perceptron (MLP) is a type of Artificial Neural Network that contains multiple layers of neurons.  
It is used to learn complex patterns from data and perform tasks such as classification and prediction.

A neural network generally contains three types of layers:

1. Input Layer
2. Hidden Layer(s)
3. Output Layer

Example inputs (as shown in the PDF example):

x1 = IQ  
x2 = Study Hours  
x3 = Play Hours  

Output = Pass or Fail

Each input is connected to neurons in the hidden layer through weights.

The general neuron computation is:

z = w1*x1 + w2*x2 + w3*x3 + ... + wn*xn + b

Where:

x = input values  
w = weights  
b = bias  
z = weighted sum

The output of the neuron is then passed through an activation function.

---

# Forward Propagation

Forward propagation means passing the input data through the network layer by layer until we get the final output.

Step 1: Compute weighted sum

z = w1*x1 + w2*x2 + w3*x3 + b

Example from the notes:

z = 95*0.01 + 4*0.02 + 4*0.03 + 1*0.01  
z = 1.151

Step 2: Apply activation function

The sigmoid activation function is used:

sigmoid(z) = 1 / (1 + e^(-z))

Example:

sigmoid(1.151) = 0.759

This value becomes the output of the hidden neuron.

Step 3: Pass the output to the next layer

Next layer calculation:

z = o1*w4 + b2

Example:

z = 0.759*0.02 + 0.03  
z = 0.04518

Activation again:

sigmoid(0.04518) = 0.51129

This becomes the final predicted output y_hat.

---

# Loss Function

The loss function measures the difference between the predicted output and the actual output.

Loss helps the model understand how wrong its prediction is.

General loss formula used in the notes:

Loss = (y - y_hat)^2

Where:

y = actual output  
y_hat = predicted output

Types of loss functions:

For Regression:
1. Mean Squared Error (MSE)
2. Mean Absolute Error (MAE)
3. Huber Loss

For Classification:
1. Binary Cross Entropy
2. Categorical Cross Entropy

The goal of training is to minimize the loss value.

---

# Backpropagation

Backpropagation is the process of sending the error backward through the network to update the weights.

Steps:

1. Calculate loss
2. Compute gradients of loss with respect to weights
3. Update weights
4. Repeat until loss is minimized

Backpropagation uses derivatives to determine how much each weight contributed to the error.

---

# Weight Update Formula

Weights are updated using gradient descent.

Weight update formula:

W_new = W_old - learning_rate * (dLoss / dW)

Where:

W_new = updated weight  
W_old = previous weight  
learning_rate = step size  
dLoss/dW = gradient of loss with respect to weight

Example meaning:

If gradient is positive → weight decreases  
If gradient is negative → weight increases

---

# Optimizers (Gradient Descent)

Optimizers are algorithms used to minimize the loss function.

Gradient Descent moves the weights in the direction that reduces the loss.

Weight update again:

W_new = W_old - learning_rate * gradient

Learning rate (η):

If learning rate is very large → training may diverge  
If learning rate is very small → training becomes very slow

Example small value:

learning_rate = 0.001

Goal of optimizer: reach the **global minimum of the loss function**.

---

# Chain Rule of Derivatives

Backpropagation uses the chain rule from calculus.

Chain rule allows us to compute derivatives in multi-layer networks.

Example:

dLoss/dW = (dLoss/dO3) * (dO3/dO2) * (dO2/dO1) * (dO1/dW)

This means:

Loss depends on output layer  
Output layer depends on hidden layer  
Hidden layer depends on input weights

So the gradient flows backward through each layer.

This process allows updating every weight in the network.

---

# Vanishing Gradient Problem

When training deep neural networks, gradients can become extremely small.

This problem is called the **vanishing gradient problem**.

When gradients become very small:

Weight update becomes very small

W_new ≈ W_old

This means learning almost stops.

This often happens when using the sigmoid activation function in deep networks.

---

# Sigmoid Activation Function

Sigmoid converts any value into a number between 0 and 1.

Formula:

sigmoid(z) = 1 / (1 + e^(-z))

Properties:

0 ≤ sigmoid(z) ≤ 1

Derivative of sigmoid:

sigmoid'(z) = sigmoid(z) * (1 - sigmoid(z))

The derivative range is very small:

0 ≤ sigmoid'(z) ≤ 0.25

Because the derivative is small, gradients shrink during backpropagation, causing the vanishing gradient problem.

---

# Example Gradient Calculation

Suppose the output neuron is:

O3 = sigmoid(z)

Where

z = w3*O2 + b3

Derivative:

dO3/dO2 = derivative_of_sigmoid(z) * w3

Since derivative_of_sigmoid(z) is small (≤0.25), repeated multiplication across layers makes gradients extremely small.

This leads to vanishing gradients.

---

# Activation Functions Used to Solve This Problem

To overcome vanishing gradients, researchers introduced better activation functions.

Common activation functions:

1. Tanh
2. ReLU
3. PReLU
4. Swish

ReLU is most commonly used in modern deep learning.

ReLU formula:

ReLU(x) = max(0, x)

Advantages of ReLU:

- Avoids vanishing gradient
- Faster training
- Simple computation

---

# Summary

A Multilayer Perceptron is a neural network containing input, hidden, and output layers.

Forward propagation calculates outputs by applying weights and activation functions.

Loss functions measure prediction error.

Backpropagation computes gradients and updates weights using gradient descent.

Chain rule helps calculate gradients across layers.

Sigmoid activation can cause the vanishing gradient problem, which is solved using newer activation functions like ReLU.
