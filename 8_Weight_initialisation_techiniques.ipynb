# <span style="color:#0B5394;"><b>1. Introduction</b></span>

In deep learning, a neural network learns by updating:

- weights
- biases

At the start of training, these weights are initialized with some values.

This process is called:

# <span style="color:#38761D;"><b>Weight Initialization</b></span>

The choice of initial weights is very important because it directly affects:

- training speed
- convergence
- model accuracy
- gradient flow

Poor initialization can make training fail.

---

# <span style="color:#0B5394;"><b>2. What is Weight Initialization?</b></span>

Weight Initialization means:

# <span style="color:#38761D;"><b>Assigning initial values to weights before training starts</b></span>

These values are usually:

- small random numbers

The network then updates them during backpropagation.

---

# <span style="color:#0B5394;"><b>3. Why Weight Initialization is Important?</b></span>

Good weight initialization helps:

1. Faster convergence
2. Stable learning
3. Better gradient flow
4. Avoid exploding gradients
5. Avoid vanishing gradients

Bad initialization may cause:

- very large gradients
- very small gradients
- unstable training
- slow learning

---

# <span style="color:#0B5394;"><b>4. Key Rules of Weight Initialization</b></span>

### Rule 1

# <span style="color:#38761D;"><b>Weights should be small</b></span>

### Rule 2

# <span style="color:#38761D;"><b>Weights should not be same</b></span>

### Rule 3

# <span style="color:#38761D;"><b>Weights should have good variance</b></span>

These rules help proper learning.

---

# <span style="color:#0B5394;"><b>5. Why Weights Should Not Be Same?</b></span>

Suppose all neurons start with same weights.

Then:

- all neurons produce same output
- all neurons learn same thing
- network loses learning power

This is called:

# <span style="color:#38761D;"><b>Symmetry Problem</b></span>

To avoid this, weights are initialized randomly.

---

# <span style="color:#0B5394;"><b>6. Exploding Gradient Problem</b></span>

Exploding gradient happens when gradients become:

# <span style="color:#38761D;"><b>Very Large</b></span>

during backpropagation.

If weights are large:

## <span style="color:#38761D;"><b>Gradient × Gradient × Gradient → Huge Value</b></span>

This causes:

- unstable training
- overflow
- huge weight updates
- model divergence

---

# <span style="color:#0B5394;"><b>7. Vanishing Gradient Problem</b></span>

Vanishing gradient is the opposite problem.

Gradients become:

# <span style="color:#38761D;"><b>Very Small</b></span>

during backpropagation.

### Result:

- early layers stop learning
- updates become almost zero
- training becomes very slow

This is common in:

- deep neural networks
- sigmoid activation
- tanh activation

---

# <span style="color:#0B5394;"><b>8. Relation Between Gradients and Weight Initialization</b></span>

The gradient update formula is:

## <span style="color:#38761D;"><b>W(new) = W(old) - η × (∂L / ∂W)</b></span>

If gradients become:

### Very Large

→ exploding gradient

### Very Small

→ vanishing gradient

So:

# <span style="color:#38761D;"><b>Proper weight initialization controls gradient flow</b></span>

---

# <span style="color:#0B5394;"><b>9. Uniform Distribution Initialization</b></span>

This is one of the simplest initialization methods.

Weights are initialized using:

# <span style="color:#38761D;"><b>Uniform Distribution</b></span>

### Formula:

## <span style="color:#38761D;"><b>Wij ∈ [ -1/√input , +1/√input ]</b></span>

Where:

- `input` = number of input neurons

### Example:

If input neurons = 3

Then:

## <span style="color:#38761D;"><b>Wij ∈ [ -1/√3 , +1/√3 ]</b></span>

---

# <span style="color:#0B5394;"><b>10. Why Uniform Initialization is Used?</b></span>

It helps:

- keep weights small
- avoid same values
- maintain controlled variance

### But:

It is not best for deep networks.

More advanced techniques perform better.

---

# <span style="color:#0B5394;"><b>11. Xavier / Glorot Initialization</b></span>

This technique was introduced by:

# <span style="color:#38761D;"><b>Xavier Glorot</b></span>

It is designed to:

# <span style="color:#38761D;"><b>Maintain stable variance across layers</b></span>

This helps avoid:

- exploding gradients
- vanishing gradients

Xavier works very well with:

- sigmoid
- tanh

activation functions.

---

# <span style="color:#0B5394;"><b>12. Xavier Normal Initialization</b></span>

Weights are initialized using:

# <span style="color:#38761D;"><b>Normal Distribution</b></span>

### Formula:

## <span style="color:#38761D;"><b>Wij ∼ N(0, σ)</b></span>

Where:

## <span style="color:#38761D;"><b>σ = √(2 / (input + output))</b></span>

Here:

- `input` = number of input neurons
- `output` = number of output neurons

---

# <span style="color:#0B5394;"><b>13. Xavier Uniform Initialization</b></span>

Weights are initialized using:

# <span style="color:#38761D;"><b>Uniform Distribution</b></span>

### Formula:

## <span style="color:#38761D;"><b>Wij ∈ [ -√(6/(input+output)) , +√(6/(input+output)) ]</b></span>

### Goal:

Maintain balanced variance across layers.

---

# <span style="color:#0B5394;"><b>14. Advantages of Xavier Initialization</b></span>

1. Reduces vanishing gradients
2. Stable training
3. Better convergence
4. Works well with sigmoid and tanh
5. Maintains variance balance

---

# <span style="color:#0B5394;"><b>15. Disadvantages of Xavier Initialization</b></span>

1. Not best for ReLU
2. Performance drops in very deep ReLU networks
3. Better alternatives exist for ReLU activations


---

# <span style="color:#0B5394;"><b>16. Kaiming He Initialization</b></span>

Kaiming He Initialization is also called:

# <span style="color:#38761D;"><b>He Initialization</b></span>

It was introduced for:

# <span style="color:#38761D;"><b>ReLU and its variants</b></span>

This technique improves training in deep neural networks using:

- ReLU
- Leaky ReLU
- ELU

activation functions.

---

# <span style="color:#0B5394;"><b>17. Why He Initialization is Needed?</b></span>

ReLU activation removes negative values.

This changes variance during forward propagation.

Xavier initialization is not enough for ReLU because:

- variance reduces layer by layer
- gradients may become weak

He initialization solves this problem by using:

# <span style="color:#38761D;"><b>Larger variance for ReLU networks</b></span>

---

# <span style="color:#0B5394;"><b>18. He Normal Initialization</b></span>

Weights are initialized using:

# <span style="color:#38761D;"><b>Normal Distribution</b></span>

### Formula:

## <span style="color:#38761D;"><b>Wij ∼ N(0, σ)</b></span>

Where:

## <span style="color:#38761D;"><b>σ = √(2 / input)</b></span>

Here:

- `input` = number of input neurons

### Main idea:

Maintain proper variance for ReLU activations.

---

# <span style="color:#0B5394;"><b>19. He Uniform Initialization</b></span>

Weights are initialized using:

# <span style="color:#38761D;"><b>Uniform Distribution</b></span>

### Formula:

## <span style="color:#38761D;"><b>Wij ∈ [ -√(6/input) , +√(6/input) ]</b></span>

This method also maintains stable variance for deep ReLU networks.

---

# <span style="color:#0B5394;"><b>20. Advantages of He Initialization</b></span>

1. Best for ReLU activation
2. Faster convergence
3. Better gradient flow
4. Reduces vanishing gradient problem
5. Works well in deep neural networks
6. Improves training stability

---

# <span style="color:#0B5394;"><b>21. Disadvantages of He Initialization</b></span>

1. Mainly designed for ReLU
2. Not ideal for sigmoid/tanh
3. Wrong usage may reduce performance

---

# <span style="color:#0B5394;"><b>22. Difference Between Xavier and He Initialization</b></span>

| Parameter | Xavier Initialization | He Initialization |
|---|---|---|
| Designed For | Sigmoid / Tanh | ReLU / Leaky ReLU |
| Variance Formula | 2 / (input + output) | 2 / input |
| Gradient Stability | Good for sigmoid/tanh | Good for ReLU |
| Deep Networks | Moderate | Excellent |
| Most Suitable Activation | Tanh | ReLU |

---

# <span style="color:#0B5394;"><b>23. Difference Between Uniform and Normal Initialization</b></span>

| Parameter | Uniform Initialization | Normal Initialization |
|---|---|---|
| Distribution Type | Uniform | Gaussian / Normal |
| Value Range | Fixed range | Around mean |
| Shape | Flat distribution | Bell-shaped distribution |
| Formula Example | [-a, +a] | N(0,σ) |
| Common Use | Xavier Uniform, He Uniform | Xavier Normal, He Normal |

---

# <span style="color:#0B5394;"><b>24. Which Initialization Should Be Used?</b></span>

### Use Xavier Initialization when:

- activation = sigmoid
- activation = tanh

### Use He Initialization when:

- activation = ReLU
- activation = Leaky ReLU

### Use Uniform Initialization when:

- simple networks
- basic experiments
