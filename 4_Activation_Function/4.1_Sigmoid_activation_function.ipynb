## 1. What is an Activation Function?

An **activation function** is a mathematical function applied to the output of a neuron in a neural network.

Purpose:
- Introduces **non-linearity**
- Allows neural networks to learn **complex patterns**
- Without activation functions, neural networks behave like **linear regression models**.

Example:
Neuron computation

z = w1x1 + w2x2 + w3x3 + b

where

w = weights  
x = inputs  
b = bias  

The activation function then converts this value into the final output.

Output = φ(z)

## 2. Sigmoid Activation Function

The **Sigmoid function** is one of the most commonly used activation functions.

Formula:

σ(x) = 1 / (1 + e^(-x))

Where:
x = input value (weighted sum of neuron inputs)

Range of output:

0 < σ(x) < 1

This means it converts any real value (-∞ to +∞) into a value between **0 and 1**.

Because of this property, it is widely used in **binary classification problems** where the output represents probability.



## 3. Graph of the Sigmoid Function

The graph of the sigmoid function is **S-shaped (logistic curve)**.

Important characteristics:

1. When x → +∞
   σ(x) → 1

2. When x → -∞
   σ(x) → 0

3. When x = 0
   σ(x) = 0.5

This means:

Large positive inputs → output close to 1  
Large negative inputs → output close to 0

Thus the function **compresses or squashes values into the range (0,1)**.



## 4. Forward Propagation with Sigmoid

In a neural network neuron:

Step 1: Compute weighted sum

z = w1x1 + w2x2 + ... + wn*xn + b

Step 2: Apply sigmoid activation

o = σ(z)

This produces the neuron output.

This process is called **Forward Propagation**.


## 5. Why Do We Need the Derivative?

During training, neural networks adjust weights using **backpropagation**.

Backpropagation uses **gradients (derivatives)** to update weights.

Error flows backward through the network and the derivative helps determine:

How much each weight should change.


## 6. Derivative of Sigmoid Function

Sigmoid function:

σ(x) = 1 / (1 + e^(-x))

Derivative:

σ'(x) = σ(x) (1 - σ(x))

This means:

Derivative = output * (1 − output)


## 7. Step-by-Step Derivation

Given:

y = σ(x) = 1 / (1 + e^(-x))

Step 1: Let

u = 1 + e^(-x)

Then

y = 1 / u

Step 2: Differentiate

dy/du = -1 / u²

Step 3: Differentiate u

du/dx = -e^(-x)

Step 4: Apply Chain Rule

dy/dx = dy/du × du/dx

dy/dx = (-1/u²) × (-e^(-x))

dy/dx = e^(-x) / (1 + e^(-x))²

Step 5: Rewrite in sigmoid form

σ'(x) = σ(x) (1 − σ(x))



## 8. Graph Interpretation

Sigmoid Graph

- Smooth S-shaped curve
- Output between 0 and 1

Derivative Graph

- Bell-shaped curve
- Maximum slope at x = 0

Important fact:

Maximum derivative value = 0.25

Meaning learning is fastest near x = 0.



```python
import numpy as np
import matplotlib.pyplot as plt

# sigmoid function
def sigmoid(x):
    return 1/(1 + np.exp(-x))

# derivative of sigmoid
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s*(1-s)

# values
x = np.linspace(-10,10,100)

# plots
plt.figure()

plt.plot(x, sigmoid(x), label="Sigmoid Function")
plt.plot(x, sigmoid_derivative(x), label="Derivative of Sigmoid")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Sigmoid Function and Its Derivative")

plt.legend()
plt.show()
```

## 10. Disadvantages of Sigmoid Function

1. Vanishing Gradient Problem

When input is very large or very small, derivative becomes close to 0.

This slows learning.

2. Not zero-centered

Outputs range from 0 to 1 instead of -1 to 1.

3. Computationally expensive compared to ReLU.
