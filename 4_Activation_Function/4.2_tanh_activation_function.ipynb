## 2. Tanh (Hyperbolic Tangent) Activation Function

The **Tanh function** is a commonly used activation function in neural networks.

It transforms input values into the range:

-1 to +1

Mathematical formula:

tanh(x) = (e^x - e^-x) / (e^x + e^-x)

Range:

-1 < tanh(x) < 1

Key idea:

Negative inputs → negative output
Positive inputs → positive output

Thus the function is **zero-centered**, which helps neural networks train more efficiently.


## 3. Output Range Comparison

| Activation | Output Range |
| ---------- | ------------ |
| Sigmoid    | 0 to 1       |
| Tanh       | -1 to 1      |

Because tanh outputs **both positive and negative values**, it is often better than sigmoid for hidden layers.

Example:

Input → Output

-10 → -1
0 → 0
10 → 1


## 4. Graph of the Tanh Function

The tanh graph has an **S-shaped curve** similar to sigmoid.

Important properties:

1. When x → +∞
   tanh(x) → 1

2. When x → -∞
   tanh(x) → -1

3. When x = 0
   tanh(x) = 0

Meaning:

Large positive input → output close to **+1**
Large negative input → output close to **-1**

This property helps keep neuron outputs balanced around zero.


## 5. Forward Propagation using Tanh

In a neural network neuron:

Step 1: Compute weighted sum

z = w₁x₁ + w₂x₂ + w₃x₃ + ... + b

Step 2: Apply activation function

o = tanh(z)

Where

w = weights
x = inputs
b = bias

The output **o** becomes the input to the next neuron.


## 6. Why Do We Need the Derivative?

During neural network training we use **Backpropagation**.

Backpropagation calculates gradients using derivatives to update weights.

Gradient tells:

How much each weight should change to reduce error.

Activation functions must be **differentiable** so gradients can be computed during training.


## 7. Derivative of Tanh Function

The derivative of tanh is:

d/dx (tanh(x)) = 1 − tanh²(x)

This means:

Derivative = 1 − (output)²

If

y = tanh(x)

then

dy/dx = 1 − y²

This formula is used in **backpropagation** to update neural network weights.


## 8. Backpropagation with Tanh

During training:

Error is propagated backwards through the network.

Using chain rule:

∂L/∂w = ∂L/∂o × ∂o/∂z × ∂z/∂w

Where

o = tanh(z)

Derivative used:

∂o/∂z = 1 − tanh²(z)

This derivative helps calculate gradients and update weights.


## 9. Graph of Derivative

Derivative graph characteristics:

* Bell-shaped curve
* Maximum slope near x = 0
* Derivative approaches 0 for large positive or negative values

Meaning:

Learning is **fastest near x = 0**.

But when inputs become very large or very small, gradients become small which may slow learning.



```python
import numpy as np
import matplotlib.pyplot as plt

# tanh function
def tanh(x):
    return np.tanh(x)

# derivative of tanh
def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# generate x values
x = np.linspace(-10, 10, 100)

# plot
plt.figure()

plt.plot(x, tanh(x), label="Tanh Function")
plt.plot(x, tanh_derivative(x), label="Derivative of Tanh")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Tanh Activation Function and Derivative")

plt.legend()
plt.show()
```

## 10. Advantages of Tanh

1. Zero-centered output
   Helps faster convergence.

2. Stronger gradients than sigmoid
   Learning can be faster.

3. Smooth and differentiable.

4. Works well in hidden layers of neural networks.


## 11. Disadvantages of Tanh

1. Vanishing gradient problem

For large positive or negative inputs, gradients become very small.

2. Computationally expensive compared to ReLU.

3. Saturation problem

Outputs near ±1 cause slow learning.
