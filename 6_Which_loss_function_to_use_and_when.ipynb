# <span style="color:#0B5394;"><b>Right Combination of Activation Function + Problem Type + Loss Function</b></span>

| Case | Hidden Layer Activation | Output Layer Activation | Problem Statement | Loss Function |
|---|---|---|---|---|
| 1 | ReLU or its Variants | Sigmoid | Binary Classification | Binary Cross Entropy |
| 2 | ReLU or its Variants | Softmax | Multi-Class Classification | Categorical Cross Entropy / Sparse Categorical Cross Entropy |
| 3 | ReLU or its Variants | Linear | Regression | MSE, MAE, Huber Loss, RMSE |

---

# <span style="color:#0B5394;"><b>Short Explanation</b></span>

**1. Binary Classification**

Used when there are only **2 classes** (Example: Yes/No, Spam/Not Spam)

- Hidden Layer: ReLU
- Output Layer: Sigmoid
- Loss Function: Binary Cross Entropy

---

**2. Multi-Class Classification**

Used when there are **more than 2 classes** (Example: Cat, Dog, Bird)

- Hidden Layer: ReLU
- Output Layer: Softmax
- Loss Function: Categorical Cross Entropy

---

**3. Regression**

Used when predicting **continuous values** (Example: House Price Prediction)

- Hidden Layer: ReLU
- Output Layer: Linear
- Loss Function: MSE / MAE / Huber Loss / RMSE

---

# <span style="color:#0B5394;"><b>Difference Between Binary, Multi-Class, and Sparse Classification</b></span>

| Type | Number of Classes | Example | Output Layer | Loss Function | Label Format |
|---|---|---|---|---|---|
| Binary Classification | Only 2 classes | Pass/Fail, Spam/Not Spam | Sigmoid | Binary Cross Entropy | 0 or 1 |
| Multi-Class Classification | More than 2 classes | Cat/Dog/Bird | Softmax | Categorical Cross Entropy | One-Hot Encoded |
| Sparse Multi-Class Classification | More than 2 classes | Cat/Dog/Bird | Softmax | Sparse Categorical Cross Entropy | Integer Labels |

---

# <span style="color:#0B5394;"><b>Detailed Explanation</b></span>

**Binary Classification**

Binary means only **two possible outputs**.

Example:

- Yes / No
- True / False
- Spam / Not Spam
- Pass / Fail

The model predicts only one probability using **Sigmoid**.

Example:

If output = 0.92

It means 92% chance of class 1.

Labels are usually:

## <span style="color:#38761D;"><b>0 or 1</b></span>

Loss function used:

## <span style="color:#38761D;"><b>Binary Cross Entropy</b></span>

---

**Multi-Class Classification**

Multi-class means there are **more than 2 classes**, but only **one class is correct**.

Example:

- Cat
- Dog
- Bird

Only one of them can be correct.

The model uses **Softmax** to give probability for each class.

Example:

## <span style="color:#38761D;"><b>[0.1, 0.8, 0.1]</b></span>

This means:

- Cat = 10%
- Dog = 80%
- Bird = 10%

Labels are stored using **One-Hot Encoding**

Example:

## <span style="color:#38761D;"><b>[0, 1, 0]</b></span>

Loss function used:

## <span style="color:#38761D;"><b>Categorical Cross Entropy</b></span>

---

**Sparse Multi-Class Classification**

This is same as multi-class classification, but labels are stored differently.

Instead of writing:

## <span style="color:#38761D;"><b>[0, 1, 0]</b></span>

we simply write:

## <span style="color:#38761D;"><b>1</b></span>

This saves memory and is easier for large datasets.

Output layer is still:

## <span style="color:#38761D;"><b>Softmax</b></span>

Loss function used:

## <span style="color:#38761D;"><b>Sparse Categorical Cross Entropy</b></span>

---

# <span style="color:#0B5394;"><b>Quick Memory Trick</b></span>

- Binary = 2 classes
- Multi-Class = Many classes + One-Hot Labels
- Sparse = Many classes + Integer Labels

---

# <span style="color:#0B5394;"><b>Simple Exam Line</b></span>

Binary classification is used when there are only two output classes and uses Sigmoid with Binary Cross Entropy.

Multi-class classification is used when there are more than two classes and uses Softmax with Categorical Cross Entropy.

Sparse categorical classification is similar to multi-class classification but uses integer labels instead of one-hot encoded labels.
